# Preprocessing of CCLE gene expression dataset

# Preprocessing of PubChem compound dataset

```python
# Read compound data from a CSV file into a DataFrame
compound_data_raw = pd.read_csv("PubChem_compound_list_ezze4ClmTNp79MTtRpWNxa3o_4jhyC9vVUo0I05bJiJOQho.csv")
```

```python
compound_data_raw
```

```python
# Extract relevant columns from compound_data_raw DataFrame
compound_data = compound_data_raw[['cid', 'isosmiles']]
```

```python
# Rename column ' cid' to 'PubCHEM' in compound_data DataFrame
compound_data = compound_data.rename(columns = {'cid':'PubCHEM'}, inplace = False)
```

```python
# Convert 'PubCHEM' column values to string type
compound_data['PubCHEM']=compound_data['PubCHEM'].astype(str)
```

```python
compound_data
```

```python
# Merge gene_drug_response_pubchem DataFrame with compound_data DataFrame on 'PubCHEM' column
gene_drug_response_isosmiles = pd.merge(gene_drug_response_pubchem, compound_data, on=['PubCHEM'], how='left')
```

```python
gene_drug_response_isosmiles
```

```python
# Drop rows with missing values (NaN) from gene_drug_response_isosmiles DataFrame
gene_drug_response_isosmiles = gene_drug_response_isosmiles.dropna(axis='index')
```

```python
gene_drug_response_isosmiles
```

```python
to_drop = ['PubCHEM']
```

```python
# Drop column 'PubCHEM' from gene_drug_response_isosmiles DataFrame
gene_drug_response_isosmiles.drop(to_drop, inplace=True, axis=1)
```

```python
# Reset the index of gene_drug_response_isosmiles DataFrame
gene_drug_response_isosmiles = gene_drug_response_isosmiles.reset_index(drop= True)
```

```python
gene_drug_response_isosmiles
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
arr = []
```

```python
# Generate Morgan fingerprints for each compound and store them in arr list
for i in range(len(gene_drug_response_isosmiles)):
    
    # Define the SMILES string for drug
    smiles = gene_drug_response_isosmiles['isosmiles'][i]
    
    # Generate Rockit molecule object from SMILES
    mol = Chem.MolFromSmiles(smiles)
    
    # Generate Morgan fingerprint with radius 2 and 256 bits
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits =256)
    
    # Convert fingerprint to a numpy array
    fp_array = np.zeros((1,), dtype = np.int64)
    
    # Display the hashed count Morgan fingerprint
    AllChem.DataStructs.ConvertToNumpyArray(fp, fp_array)
    arr.append(fp_array)
```

```python
# Convert the list of Morgan fingerprints into a DataFrame
morgan_data =pd.DataFrame(arr)
```

```python
morgan_data
```

```python
# Join morgan_data DataFrame with gene_drug_response_isosmiles DataFrame
gene_drug_response_morgan = gene_drug_response_isosmiles.join(morgan_data)
```

```python
gene_drug_response_morgan
```

```python
to_drop = ['isosmiles']
```

```python
# Drop column 'isosmiles' from gene_drug_response_morgan DataFrame
gene_drug_response_morgan.drop(to_drop, inplace=True, axis=1)
```

```python
gene_drug_response_morgan
```

```python
# Extract 'LN_IC50' column values to list
LN_IC50_lst = gene_drug_response_morgan['LN_IC50'].values.tolist()
```

```python
to_drop = ['LN_IC50']
```

```python
# Drop column 'LN_IC50' from gene_drug_response_morgan DataFrame
gene_drug_response_morgan.drop(to_drop, inplace=True, axis=1)
```

```python
# Insert 'LN_IC50' column at index 19481 in gene_drug_response_morgan DataFrame
gene_drug_response_morgan.insert(loc = 19481,
          column = 'LN_IC50',
          value = LN_IC50_lst)
```
